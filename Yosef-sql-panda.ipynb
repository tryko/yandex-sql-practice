#  Table Queries <a class="anchor" id="chapter3"></a>

function that takes a query and return dataframe for general use

```python
def queryResult(q):
    return pd.io.sql.read_sql(q, con = engine)
```

```python
from sqlalchemy import inspect
inspector = inspect(engine)

tables = inspector.get_table_names()
```

1. What is the position played by the majority of the squad?

```python
q='''
select count(*),position from squad group by position order by count desc limit 1
'''
queryResult(q)
```

2. Which players played the most games during the entire season?

```python
q='''
select squad.player_id,first_name,last_name,sum(league+champions+cup+supercup) as total_games 
from games_played
inner join squad on games_played.player_id=squad.player_id
group by squad.player_id,first_name,last_name 
order by total_games desc
'''
queryResult(q)
```

<div class="alert alert-success" role="alert">
  Great!
</div>

3. Out of the players who played less than 1000 minutes in the league who scored the most goals?

```python
q ='''
select squad.player_id, first_name,last_name,league_goals,league_minutes 
from league_stats 
inner join squad on league_stats.player_id=squad.player_id
where league_minutes<1000 order by league_goals desc 
limit 1
'''
queryResult(q)
```

  Great!
</div>

4. Which players played the most minutes in the league? Show top 5.

```python
q ='''
select league_stats.player_id,first_name,last_name,league_minutes
from league_stats 
inner join squad on league_stats.player_id=squad.player_id
order by league_minutes desc limit 5

'''
queryResult(q)
```

5.How many assists were made by the forwards?

```python
q ='''
select sum(assists) as assists_by_fw from
(select squad.player_id,
sum(league_assists+champions_assists+cup_assists+supercup_assists) as assists
from squad 
inner join league_stats on squad.player_id=league_stats.player_id
inner join champions_stats on squad.player_id= champions_stats.player_id
inner join cup_stats on squad.player_id= cup_stats.player_id
inner join supercup_stats on squad.player_id= supercup_stats.player_id
where position='fw' 
group by squad.player_id) as x
'''
queryResult(q)
```

<div class="alert alert-success" role="alert">
  Great!
</div>

6. Who are the top 3 players with the best goals per minute record?

```python
q='''
select *, 
case when x.sum_minutes=0 or  x.sum_goals=0 then 0
else cast(x.sum_minutes as decimal)/cast(x.sum_goals as decimal) end as goal_per_minute
from 
(select squad.player_id,first_name,last_name,
sum(league_minutes+champions_minutes+cup_minutes+supercup_minutes) as sum_minutes,
sum(league_goals+champions_goals+cup_goals+supercup_goals) as sum_goals
from squad 
inner join league_stats on squad.player_id=league_stats.player_id
inner join champions_stats on squad.player_id= champions_stats.player_id
inner join cup_stats on squad.player_id= cup_stats.player_id
inner join supercup_stats on squad.player_id= supercup_stats.player_id
group by squad.player_id,first_name,last_name) as x
order by goal_per_minute desc limit 3

'''
queryResult(q)
```

9. Which defender contribute the most goals and assists in the league?

```python
q='''
select squad.player_id,first_name,last_name,
sum(league_goals+league_assists) as goals_assists
from squad
inner join league_stats on squad.player_id=league_stats.player_id
where position='df'
group by squad.player_id,first_name,last_name
order by goals_assists desc 
limit 1

'''

queryResult(q)
```

10. How many players played more then 50 hours total in the league and in the champions league?

```python
q='''
select count(*) as players_more_than_50_hours from 
(select sum(league_minutes+champions_minutes) from champions_stats 
inner join league_stats
on champions_stats.player_id=league_stats.player_id
group by league_stats.player_id
having sum(league_minutes+champions_minutes)>50*60) as x

'''

queryResult(q)
```
